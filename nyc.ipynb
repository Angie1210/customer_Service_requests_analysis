```python
def convert_lowercase(df):
    #convert string columns into lowercase to elimintae duplication
    df = df.applymap(lambda s:s.lower() if type(s) == str else s)
    return df
```

```python
df = convert_lowercase(df)
df
```

```python
#drop columns that are almost empty
def drop_empty_columns(df,threshold=.085):
    #col with missing values
    col_count=df.isna().sum().where(lambda x:x>0).dropna()
    #cols to be drop
    
    columns_to_drop = []
    
    #iterate through columns wih missing values
    for col,miss_values in col_count.iteritems():
        
        if (miss_values/len(df) >= threshold):
            columns_to_drop.append(col)
            
    df.drop(columns= columns_to_drop,axis=1,inplace=True)

    return df
```

```python
df=drop_empty_columns(df,threshold=.085)
df
```
